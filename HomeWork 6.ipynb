# Loading of historical data for S&P 500 Index since Jan, 1, 2020,

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import yfinance as yf

SPY = yf.download('SPY', start="2020-01-01")
SPY.head(3)
```

# 1. A plot that shows the price change in 2020 (use the close price)

```python
from datetime import datetime
fig = plt.figure(figsize=(10, 10))
axe = fig.add_subplot(2, 2, 1) # we add axe for the subplot of S&P 500 Index

# plot SPY price change (close price in 2020)
axe.plot(SPY.index, SPY['Close'])
axe.set_title("S&P 500 Index")
axe.set_ylim([200, 400])
plt.xlim([datetime(2020, 1, 1), datetime(2020, 12, 1)])
ticks = [datetime(2020, x, 1) for x in range(1, 13)]
#ticks.append(datetime(2020, 1, 11))
plt.xticks(ticks, rotation=45)
plt.show()
```

# A. Annotations for the highest value before COVID and the lowest value after that.

```python
from datetime import timedelta

plt.figure(figsize=(10, 5))
plt.plot(SPY.index, SPY['Close'])

# Point out the peak value before COVID
plt.annotate('Peak value: 338',
             xy=(datetime(2020, 2, 20), 338),
             xytext=(datetime(2020, 2, 20)+timedelta(days=15), 340),
             arrowprops=dict(facecolor='red', headwidth=4, width=4, headlength=4))

# Let point out the lowest value after that
plt.annotate('Lowest value: 222',
             xy=(datetime(2020, 3, 23), 222),
             xytext=(datetime(2020, 3, 23)+timedelta(days=15), 225),
             arrowprops=dict(facecolor='blue', headwidth=4, width=4, headlength=4))
```

# B. A rectangle indicating how long it takes for the price to drop from peak to bottom after the COVID outbreak.

```python
from datetime import timedelta

plt.figure(figsize=(10, 5))
plt.plot(SPY.index, SPY['Close'])

# Point out the peak value before COVID
plt.annotate('Peak value: 338',
             xy=(datetime(2020, 2, 20), 338),
             xytext=(datetime(2020, 2, 20)+timedelta(days=15), 340),
             arrowprops=dict(facecolor='red', headwidth=4, width=4, headlength=4))

# Let point out the lowest value after that
plt.annotate('Lowest value: 222',
             xy=(datetime(2020, 3, 23), 222),
             xytext=(datetime(2020, 3, 23)+timedelta(days=15), 225),
             arrowprops=dict(facecolor='blue', headwidth=4, width=4, headlength=4))


# indication of the rectangle
rectangle = plt.Rectangle((datetime(2020, 2, 18), 200),
                          timedelta(days=40),
                          150,
                          color='k',
                          alpha=0.3)
plt.gca().add_patch(rectangle)
```
