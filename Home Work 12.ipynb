```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Find the top 5 countries with the largest amount of large airports.
# let load those datasets
airports = pd.read_csv('airports.csv')
airport_frequencies = pd.read_csv('airport-frequencies.csv')
countries1 = pd.read_csv('countries1.csv')
regions = pd.read_csv('regions.csv')
```

```python
# let load airports data 
airports.head()
```

```python
airports.shape
```

```python
# let extract all large airports 
airports_large = airports[airports['type'] == 'large_airport']
airports_large.head()
```

```python
airports_large.shape
```

```python
airport_frequencies.head()
```

```python
# let extract the frequency of all large airports
frq_large_airports = pd.DataFrame()

for ident in airports_large['ident']:
    frq1 = airport_frequencies[airport_frequencies['airport_ident'] == ident]
    #print(frq1)
    frq_large_airports = pd.concat([frq_large_airports, frq1])
    
frq_large_airports
    
```

```python
countries1.head()
```

```python
# let extract the countries of all large airports
countries_large_airports = pd.DataFrame()

for country_code in countries1['code']:
    country = airports_large[airports_large['iso_country'] == country_code]
    countries_large_airports = pd.concat([countries_large_airports, country])

    
countries_large_airports
```

```python
# Find the top 5 countries with the largest amount of large airports.
countries = countries_large_airports.groupby('iso_country')
results = countries.size()
df = results.to_frame(name = 'num_large_airport')
df.head()
```

```python
# let add countries name columns
DataFrame = pd.merge(df, countries1[['code', 'name']], left_index = True, right_on = 'code')
DataFrame.head()
```

```python
# Let sort in descending order the number of large airoprt
Top_5_Countries = DataFrame.sort_values(by = 'num_large_airport', ascending = False).reset_index(drop=True)
Top_5_Countries.head()
```

```python
# 2.Create a data frame with columns "Name", "Code", "Number of Large Airports", and "Numberof All Airports".
# Let first create a data frame with the number of all airports
countries = airports.groupby('iso_country')
results = countries.size()
df1 = results.to_frame(name = 'Number of All Airports')
df1.head()
```

```python
# Let combine the data df of number of large airports and df1 of Number of all airports
data = pd.merge(df, df1, on = 'iso_country')
data.head()
```

```python
New_Data_Frame = pd.merge(data, countries1[['code', 'name']], left_index = True, right_on = 'code')
New_Data_Frame.head()
```
