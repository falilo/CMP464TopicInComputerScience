```python
import numpy as np
import pandas as pd
```

```python
# Reference:
# # https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
url = "https://shanelynnwebsite-mid9n9g1q9y8tt.netdna-ssl.com/wp-content/uploads/2015/06/phone_data.csv"
data = pd.read_csv(url, sep=",", index_col='index')
print(data.shape)
data.head(3)
```

```python
# 1. Find out all network names that are associated to network_type "mobile".
network_names = data[data['network_type'] == 'mobile']
network_names.head()
```

```python
# 2. How many messages were sent to each mobile network every month?

# First let extract sms sent to each mobile network from network_names data 
sms_data = network_names[network_names['item'] == 'sms']

# second Split sms according to the month and network
groups = sms_data['duration'].groupby([sms_data['month'], sms_data['network']])

# third apply size function to get number of sms 
groups.size().to_frame(name = 'Number of sms') 
```

```python
# 3. What is the total call duration to each mobile network every month?
# First let extract all calls durations to each mobile network from network_names data 
call_data = network_names[network_names['item'] == 'call']

# second Split call according to the month and network
groups = call_data['duration'].groupby([call_data['month'], call_data['network']])

# third apply sum function to get total call duration to each mobile network every month
groups.sum().to_frame(name = 'Total call duration') 
```
