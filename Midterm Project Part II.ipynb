# VI. Apply the kNN (k=5) method to predict sale price of the first instance from the test set.

```python
# 1. Load test.csv and extract its first row. Calculate its total area and area per room.
url = 'https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=test1.csv'
test_data = pd.read_csv('test1.csv')
test_data.head()
```

```python
# Extract the first row
test_data_1 = test_data.loc[[0], :]
test_data_1
```

```python
#2. Add a new column (“Diff”) to the data frame representing the training data
# First let calculate total area = total area above ground (“GrLivArea”) + total basement area (“TotalBsmtSF”)
TotalArea = test_data['GrLivArea'] + test_data['TotalBsmtSF']
TotalArea.head()
```

```python
# 2nd area per room = total area above ground (“GrLivArea”) / number of rooms (“TotRmsAbvGrd”)
AreaPerRoom = test_data['GrLivArea'] / test_data['TotRmsAbvGrd']
AreaPerRoom.head()
```

```python
# let extract the relevant commun featuresfrom test data frame
relevant_features_test = ['OverallQual','YearBuilt','TotalBsmtSF','GrLivArea','TotRmsAbvGrd']
test_data = test_data.filter(relevant_features_test, axis=1)
test_data.head()
```

```python
# let create a fuction that will generate all values of Total Area 

def TotalArea (test_data):
    
    return test_data['GrLivArea'] + test_data['TotalBsmtSF']
```

```python
TotalArea(test_data.loc[0, :])
```

```python
# Let apply a column TotalArea to the test_data frame 
test_data['TotalArea'] = test_data.apply(TotalArea, axis = 1)
test_data.head()
```

```python
# let create a fuction that will generate all values of area per room 

def AreaPerRoom (test_data):
    
    return  test_data['GrLivArea'] / test_data['TotRmsAbvGrd']
```

```python
AreaPerRoom(test_data.loc[0, :])
```

```python
# let add the column AreaPerRoom to the test_data frame
test_data['AreaPerRoom'] = test_data.apply(AreaPerRoom, axis = 1)
test_data.head()
```

```python
test_data_1 = test_data.loc[[0], :]
test_data_1
```

```python
#2. Add a new column (“Diff”) to the data frame representing the training data. 
#For each row, calculate a weighted sum of the differences between this row’s features and those from the test row.
#For each feature, use the reciprocal of its standard deviation as its weight.
# data['Diff'] = diff * 1/data.std()
    
data['Diff'] = np.abs(data['OverallQual'] - test_data['OverallQual']) * (1/data['OverallQual'].std()) + \
               np.abs(data['YearBuilt'] - test_data['YearBuilt']) * (1/data['YearBuilt'].std())+ \
               np.abs(data['TotalBsmtSF'] - test_data['TotalBsmtSF']) * (1/data['TotalBsmtSF'].std()) + \
               np.abs(data['GrLivArea'] - test_data['GrLivArea']) * (1/data['GrLivArea'].std()) + \
               np.abs(data['TotRmsAbvGrd'] - test_data['TotRmsAbvGrd']) * (1/data['TotRmsAbvGrd'].std()) + \
               np.abs(data['TotalArea'] - test_data['TotalArea']) * (1/data['TotalArea'].std()) + \
               np.abs(data['AreaPerRoom'] - test_data['AreaPerRoom']) * (1/data['AreaPerRoom'].std())

data.head()
           
```

```python
# 3. Sort the rows so that values in the “Diff” column is listed in ascending order.
#The top 5 instances are the closest neighbors of the new instance y. 
# k = 5
data = data.sort_values(by=['Diff'])
data.head()
```

```python
# 4. Calculate the average sale price of these 5 instances. This will be the prediction of the new instance.
# we know the indeces of these 5 instances index = [1050,931,1449,857,950]
# Let extract the SalePrice of these 5 instances

New_data_frame[['SalePrice']].iloc[[1050,931,1449,857,950]]
```

```python
# Calculate the average sale price of these 5 instances. This will be the prediction of the new instance
prediction_new_instance = New_data_frame[['SalePrice']].iloc[[1050,931,1449,857,950]].mean()
print("Prediction:", prediction_new_instance)
```
