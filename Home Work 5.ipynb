```python
# let load movie.csv file from ml-latest-small
movies_df = pd.read_csv('Data/ml-latest-small/movies.csv', sep = ',')
movies_df.head()
```

 # 2. Find the average user rating and the number of ratings for each comedy movie.

```python
# let load ratings processed.csv file ml-latest-small
ratings_df = pd.read_csv('Data/ml-latest-small/ratings_processed.csv', sep = ',', index_col=0)
ratings_df.head()
```

```python
# 2 let extract movie Id where genres contains comedy
movies_id_comedy = (movies_comedy['genres']== 'Comedy')
print(movies_id_comedy)
```

```python
#let create the function that calculate mean of user rating movies genres comedy

def get_avg_user_rating_comedy(ratings_df, movies_id_comedy):
    
    filter_movie_comedy = (ratings_df['movieId']== movies_id_comedy)
    ratings_movie_comedy = ratings_df[filter_movie_comedy]
    
    average_user_rating_comedy = ratings_movie_comedy['rating'].mean()
    
    return average_user_rating_comedy
```

```python
get_avg_user_rating_comedy(ratings_df, 61)
```

```python
# average user rating for each comedy movie
# let create a for loop that looks in all rows of movie and select only for movie Id contains genres comedy
for idx in movies_df.index:
    
    movies_id_comedy = movies_df.loc[idx, 'movieId']
    avg_user_rating_comedy = get_avg_user_rating_comedy(ratings_df, movies_id_comedy)
    movies_df.loc[idx, 'AvgRating'] = avg_user_rating_comedy
    
movies_df.head()
```

```python
# let extract only movies genres comedy and average rating
filter_comedy = (movies_df['genres'] == 'Comedy')
movies_comedy_only = movies_df[filter_comedy]
movies_comedy_only.head()
```

```python
# number of ratings of each comedy
# let create the function that return to the number of rating of each movie comedy
def get_num_rating(ratings_df, movies_id_comedy):
    
    filter_movie_comedy = (ratings_df['movieId']== movies_id_comedy)
    ratings_movie_comedy = ratings_df[filter_movie_comedy]
    
    num_ratings_movie_comedy = ratings_movie_comedy.shape[0]
    
    return num_ratings_movie_comedy
```

```python
get_num_rating(ratings_df,5)
```

```python
#let creat a for loop that look in all rows and add a new column for number of rating 
for idx in movies_df.index:
    movies_id_comedy = movies_df.loc[idx, 'movieId']
    num_ratings_movie_comedy = get_num_rating(ratings_df, movies_id_comedy)
    movies_df.loc[idx, 'NumOfRatings'] = num_ratings_movie_comedy
    
movies_df.head()
```

```python
# let extract the number of rating for each movie genres comedy
filter2_comedy = (movies_df['genres'] == 'Comedy')
movies_comedy_only2 = movies_df[filter2_comedy]
movies_comedy_only2.head()
```
